### Matplot For visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
# lets change size of graph using figure
plt.figure(figsize=(6,2))
# alpha controls colors intesity 
plt.scatter(x,y,c="plum",alpha=.9,marker='x')
# use give label to x-axis
plt.xlabel("This is my x-axis")
# use to give label to y-axis
plt.ylabel("This is my y-axis")
# use to give title to graph
plt.title("This is my x versus y scatter graph")
# represent graph in grid
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,c='orange',marker='x')
```

```python
plt.plot(x,y,'--r')
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel('This is x-axis')
plt.ylabel('This is y-axis')
plt.title('This sin waves graph')
plt.grid()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
# plt.bar(x,y,color="orange")
plt.barh(x,y,color="orange")
plt.title('Categorical data graph')
plt.xlabel('This is category')
plt.ylabel('This is values for all category')
plt.grid()
```

```python
x = [3,4,5,6,7,8]
y = [4,5,6,7,8,9]
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,c='green')
plt.scatter(x,y,c='orange')
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color="violet")
plt.xlabel('Representing unique values.')
plt.ylabel('Representing frequencies.')
plt.grid()
# use to save out graph
plt.savefig('hist.png')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(x,y,z,c='plum')
plt.xlabel('This is x-axis')
plt.ylabel('This is y-axis')
plt.show()
```
